```python
import spacy

def extract_names(text):
    # Load spaCy English model
    nlp = spacy.load("en_core_web_sm")

    # Process the text using spaCy
    doc = nlp(text)

    # Extract names from the named entities
    names = [ent.text for ent in doc.ents if ent.label_ == "PERSON"]

    return names


```

```python
text_example = "Hi I am Harsha. I have opened an account at Chase"
extracted_names = extract_names(text_example)

print("Extracted Names:", extracted_names)
```

```python
import spacy
import re

def extract_name_and_numbers(text):
    nlp = spacy.load("en_core_web_sm")

    doc = nlp(text)

    # Extract names
    names = [ent.text for ent in doc.ents if ent.label_ == "PERSON"]

    # Extract account numbers
    account_numbers = [token.text for token in doc if token.is_digit and len(token.text) == 9]

    # regex
    phone_numbers = re.findall(r"\b(?:\+?1?[-.\s]?)?\(?\d{3}\)?[-.\s]?\d{3}[-.\s]?\d{4}\b", text)

    return names, account_numbers, phone_numbers

# Example usage
text_example = "Hi, I am Tony Stark. He is my friend Dave Dan. My account number is 110112363, and my phone number is +1 (555) 123-4567"
extracted_names, extracted_account_numbers, extracted_phone_numbers = extract_name_and_numbers(text_example)

print("Extracted Names:", extracted_names)
print("Extracted Account Numbers:", extracted_account_numbers)
print("Extracted Phone Numbers:", extracted_phone_numbers)
```

```python
import spacy
from datetime import datetime, timedelta

def extract_days_offset(text):
    # Load spaCy English model
    nlp = spacy.load("en_core_web_sm")

    # Process the text using spaCy
    doc = nlp(text)

    # Extract relevant information using spaCy's dependency parsing
    for i, token in enumerate(doc):
        if token.text.lower() == "days" and i > 0 and doc[i - 1].is_digit:
            # Extract the number of days
            days_offset = int(doc[i - 1].text)

            # Check if it's "before" or "after"
            if i + 1 < len(doc) and doc[i + 1].text.lower() in ["before", "ago"]:
                days_offset *= -1  # If "before," make the offset negative

            # Calculate the date
            calculated_date = datetime.now() + timedelta(days=days_offset)

            # Format the result
            formatted_date = calculated_date.strftime("%Y-%m-%d")

            return f"{days_offset} days from {formatted_date}"

    return "Unable to extract information from the text."

# Example
text_example_before = "What is the date 2 days before?"
text_example_after = "What is the date 2 days after?"
ex = "What is the date 2 days from now?"
ex1 = "What is the date 2 days ago?"

result_before = extract_days_offset(text_example_before)
result_after = extract_days_offset(text_example_after)
result = extract_days_offset(ex)
result1 = extract_days_offset(ex1)

print('current:', datetime.now())
print(result_before)
print(result_after)
print(result)
print(result1)
```
